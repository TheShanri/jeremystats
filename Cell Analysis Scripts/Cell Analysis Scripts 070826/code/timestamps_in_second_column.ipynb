```python
import os
import numpy as np
import pandas as pd
from scipy.io import savemat

# ========== USER INPUT ==========
kilosort_path = 'D:\Data\PTEN Project\PTEN_DKO\M61\m61s7\2025-07-11_15-53-22\binary\kilosort_tl6_tu7_fs30000.0invert_False'  # <-- Replace with actual path
Fs = 30000  # Sampling frequency in Hz
output_dir = os.path.join(kilosort_path, 'cluster_timestamps_mat')
os.makedirs(output_dir, exist_ok=True)

# ========== LOAD DATA ==========
kslabel_path = os.path.join(kilosort_path, 'cluster_KSLabel.tsv')
spike_clusters = np.load(os.path.join(kilosort_path, 'spike_clusters.npy'))
spike_times = np.load(os.path.join(kilosort_path, 'spike_times.npy'))

# ========== GET GOOD CLUSTERS ==========
labels_df = pd.read_csv(kslabel_path, sep='\t')
good_clusters = labels_df[labels_df['KSLabel'] == 'good']['cluster_id'].to_numpy()

# ========== LOOP THROUGH CLUSTERS ==========
for cluster_id in good_clusters:
    indices = np.where(spike_clusters == cluster_id)[0]
    timestamps_sec = (spike_times[indices] / Fs).reshape(-1, 1)  # column vector

    var_name = f"timestamp_{cluster_id}"
    mat_path = os.path.join(output_dir, f"{var_name}.mat")
    savemat(mat_path, {var_name: timestamps_sec})

print(f"✅ Saved {len(good_clusters)} .mat files in: {output_dir}")
```
